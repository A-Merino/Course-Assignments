```python
def alice_html():
    try:
        response = requests.get("https://www.gutenberg.org/files/11/old/alice30.txt")
        return response.content.decode()
    except Exception as e:
        print(e)


def alice():
    source = alice_html()

    with open("alice30.txt", "w") as alice_file:
        alice_file.write(source)

    story_words = read_words("alice30.txt", "*END*")
    print(f"{'CHAPTER':20}{'CHAPTER TITLE':40}{'UNIQUE WORD COUNT'}")
    for chapter in story_words.items():
        print(f"{chapter[0].split(': ')[0]:20}{chapter[0].split(': ')[1]:40}{len(chapter[1]):^15}")


alice()
```

## Question 2

### Part A

```python
def fix_sent(sent_words, very_dict, part):
    new_sentence = []
    after_very = False
    last_word = ""
    for i, word in enumerate(sent_words):
        if word.lower().find('very') != -1:
            after_very = True
            last_word = word
            continue
        if after_very and very_dict.get(re.sub(r'[^\w]',"",word).lower()) is not None:
            if part:
                word = "".join(("[",very_dict.get(re.sub(r'[^\w]',"",word).lower()).lower(),"]"))
            else:
                word = very_dict.get(re.sub(r'[^\w]',"",word).lower()).lower()
            after_very = False
            new_sentence.append(word)
        elif after_very:
            after_very = False
            new_sentence.append(last_word)            
            new_sentence.append(word)
        else:
            new_sentence.append(word)

        last_word = word
    return" ".join(new_sentence)
```

The fix_sent() function takes in the sentence as a list of strings and a dictionary to compare against and a boolean to see which part of the question it is being used for False if part A and True if part B
Inside the function:
* The loop goes through every word and checks if the word is very
* If it finds very then it goes to the next word and uses regex to check if word is in the dictionary as a key
* If it finds it in the dictionary then it will change the current word to the improved word
* At the end of every loop it adds the current word to the new sentence unless the word is very and the next word is in the dictionary
* The function then returns the sentence as a whole using the join function

```python
def q2a():

    very_dict = dict()

    with open("very_words.txt", "r") as very_file:
        line = very_file.readline()
        while line != "":
            very_dict[line.split()[1]] = line.split()[3]
            line = very_file.readline()

    very_file.close()
    sentence = input("Type a sentence to improve your vocabulary: ")
    print(fix_sent(sentence.split(), very_dict, False))



q2a()
```

* It creates the dictionary of improved words from the txt file given and runs it through the fix_sent() function with user input

```python
def q2b():

    very_dict = dict()
    with open("very_words.txt", "r") as very_file:
        line = very_file.readline()
        while line != "":
            very_dict[line.split()[1]] = line.split()[3]
            line = very_file.readline()

    very_file.close()

    with open("very_sentences.txt", "r") as sent_file:
        count = 0
        corr_count = 0
        line = sent_file.readline()
        corrected = False
        while line != "":
            count = count + 1
            if line.find(fix_sent(line.split(), very_dict, True)) == -1:
                corr_count = corr_count + 1
                print("Input:", line.strip())
                print("Output:", fix_sent(line.split(), very_dict, True))
                print()
            line = sent_file.readline()

        print(corr_count, "/", count, "sentences were corrected.")
    sent_file.close()

q2b()
```

## Question 3

```python
def q3():
    countries = dict()
    # encoding needed to read special characters
    with open("IMDb_movies.csv", newline='', encoding="utf8") as movies_file:
        file_reader = reader(movies_file)
        for row in file_reader:
            creators = row[7].split(', ')
            for c in creators:
                if countries.get(c) is None:
                    countries[c] = 1
                else:
                    countries[c] = countries[c] + 1

        sort = dict(sorted(countries.items(), key=lambda i:i[1], reverse=True))
        for i, country in enumerate(sort):
            print(f"{country:15}{'█'*int(sort[country]/1000)}{'  ('}{sort[country]/1000:.2f}{'k)'}")
            if i == 9: # To know to stop at 10 countries
                break
 

q3()
```

* The function opens the movies csv and creates a reader for it
* It goes through each row in the file and splits it if there are more than one countries that made the movie. 
* Then going through the countries that made a specific movie it creates a new entry in the dictionary with the value of 1
* If the country already has a key entry then we update the value and add one.
* Then we sort the dictionary by the values in descending order
* Then for the first 10 entries we print out the █ for however many 1 thousand movies were made
